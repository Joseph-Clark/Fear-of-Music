```python
def plot_sales(form): 
  plt.plot(form)
```

```python
digital_formats = ['Download Single', 'Download Album', 'Kiosk', 'Download Music Video',
       'Ringtones & Ringbacks', 'Paid Subscriptions']



for form in digital_formats: 
    form_sales = sales.ix[form][27:]
    plot_sales(form_sales)
  
plt.title("US Recording Industry Digital Sales Volume (2000-2015)")
plt.xlabel("Year")
plt.legend(digital_formats)
plt.legend(loc=2,prop={'size':6})
plt.ylabel('Units (millions)')
if SAVING: 
  plt.savefig("./Plots/Sales/US_Sales_digital_1.png")
plt.show()
```
