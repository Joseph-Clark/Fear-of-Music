```python
import billboard
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import collections
```

```python
SAVING = False 
```

```python
rev = pd.read_csv("./Data/US_Rev_by_Format.csv", index_col=0)
#Need to Take out $ from CSV
plt.plot(rev.ix["TOTAL"])
plt.title("US Recording Industry Total Revenue")
plt.xlabel("Year")
plt.ylabel('Inflation adjusted millions of 2015 USD')
if SAVING: 
  plt.savefig("./Plots/US_REV.png") 
plt.show()
```

```python
rev_pct_change = rev.ix["TOTAL"].pct_change()*100
rev_pct_change[0] = 0
rev_pct_change
plt.plot(rev_pct_change)
plt.title("Percent Change in US Recording Industry Revenue")
plt.xlabel("Year")
plt.ylabel('Percent change in revenue (adjusted 2015 USD)')
plt.axhline(y=0, xmin=0, xmax=1000, linewidth=1, color = 'k')
if SAVING:
  plt.savefig("./Plots/pct_change_rev1.png") 
plt.show()
```

```python
plt.plot(rev.ix["TOTAL"][27:])
plt.title("US Recording Industry Total Revenue (2000-2015)")
plt.xlabel("Year")
plt.ylabel('Inflation adjusted millions of 2015 USD')
if SAVING: 
  plt.savefig("./Plots/US_REV_2.png") 
plt.show()
```

```python
rev_pct_change_2 = rev.ix["TOTAL"][26:].pct_change()*100 #26 not 27 because stored as DATAFRAME
rev_pct_change[0] = 0
plt.plot(rev_pct_change_2)
plt.title("Percent Change in US Recording Industry Revenue (2000-2015)")
plt.xlabel("Year")
plt.ylabel('Percent change in revenue (adjusted 2015 USD)')
plt.axhline(y=0, xmin=0, xmax=1000, linewidth=1, color = 'k')
if SAVING:
  plt.savefig("./Plots/pct_change_rev2.png") 
plt.show()
```

```python
sales = pd.read_csv("./Data/US_Sales_by_Format.csv", index_col=0)
plt.plot(sales.ix["CD"])
plt.title("US Recording Industry Sales Volume (CDs)")
plt.xlabel("Year")
plt.ylabel('Units (millions)')
if SAVING: 
  plt.savefig("./Plots/US_Sales_cd.png")
plt.show()
```

```python
#2000 to present

plt.plot(sales.ix["CD"][27:])
plt.title("US Recording Industry Sales Volume (CDs)")
plt.xlabel("Year")
plt.ylabel('Units (millions)')
if SAVING: 
  plt.savefig("./Plots/US_Sales_cd_2.png")
plt.show()
```

```python
digital_formats = ['Download Single', 'Download Album', 'Kiosk', 'Download Music Video',
       'Ringtones & Ringbacks', 'Paid Subscriptions']

digital_sales = 0

for form in digital_formats: 
    form_sales = sales.ix[form][27:]
    digital_sales += form_sales
  
plt.plot(digital_sales)
plt.title("US Recording Industry Digital Sales Volume (2000-2015)")
plt.xlabel("Year")
plt.ylabel('Units (millions)')
if SAVING: 
  plt.savefig("./Plots/US_Sales_digital_1.png")
plt.show()
```

```python
#BAD because it's real price
real_unit_price_CD = rev.ix["CD "][27:]/sales.ix["CD"][27:]
plt.plot(real_unit_price_CD)
plt.title("Avg Real Price of CD (2000-2015)")
plt.xlabel("Year")
plt.ylabel('2015 USD')
if SAVING:
  plt.savefig("./Plots/US_Avg_cd_Price2.png") 
plt.show()
```

```python
digital_rev = 0 

for form in digital_formats: 
    form_sales = rev.ix[form][27:]
    digital_rev += form_sales

plt.plot(digital_rev)
plt.title("US Recording Industry Digital Sales Revenue (2000-2015)")
plt.xlabel("Year")
plt.ylabel('2015 USD (millions)')
if SAVING: 
  plt.savefig("./Plots/US_Digital_Rev_1.png")
plt.show()
```
