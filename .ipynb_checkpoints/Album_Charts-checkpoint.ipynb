```python
import billboard
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import collections
```

```python
SAVING = False 
```

```python
all_charts = {}

for i in range(0, 43):
  year = str(i + 1973)
  path = "./Data/charts/billboard-200/" + year + "_billboard-200.csv"
  df = pd.read_csv(path)
  all_charts[year] = df

charts_over_time = collections.OrderedDict(sorted(all_charts.items()))
```

```python
# GET NUMBER OF UNIQUE SONGS BY YEAR 

yr = 1973
yrs = []
unique_albums = []

for year in charts_over_time: 
  #create years list
  yrs.append(yr) 
  df = charts_over_time[year]
  unique_albums.append(len(df.columns))
  yr +=1 
```

```python
#PLOT NUMBER OF UNIQUE Albums BY YEAR

plt.plot(yrs, unique_albums)
plt.title("Number of Unique Albums on Billboard 200 Charts by year")
plt.xlabel("Year")
plt.ylabel('Number of Unique Albums on Charts')
if SAVING:
  plt.savefig("./Plots/Unique_Albums_bboard-200_1.png") 
plt.show()
```

```python
#Plot percentage change in unique songs

unique = pd.Series(unique_albums, index=yrs).pct_change()
unique[1973]=0

plt.plot(unique)
plt.title("Percent Change in Unique Albums on Billboard 200 Charts by year")
plt.xlabel("Year")
plt.ylabel('Percent change in unqiue Albums per year')
plt.axhline(y=0, xmin=0, xmax=1000, linewidth=1, color = 'k')
if SAVING:
  plt.savefig("./Plots/pct_change_unique_albums1.png") 
plt.show()
```

```python
avgs = []

for year in charts_over_time: 
  df = charts_over_time[year]

  names = list(df.columns.values)
  sum = 0 
  for song in names: 
    sum += df[song].sum()

  avg = sum/len(names)
  avgs.append(avg)
```

```python
plt.plot(yrs, avgs)
plt.title("Average Lifespan of Album on Billboard 200 Charts by year")
plt.xlabel("Year")
plt.ylabel('Average Number of Weeks on Charts')
if SAVING: 
  plt.savefig("./Plots/Avg_Weeks_Albums2.png")  
plt.show()
```

```python
#Plot percentage change in average time on charts

time_on_charts = pd.Series(avgs, index=yrs).pct_change()
time_on_charts[1973]=0

plt.plot(time_on_charts)
plt.title("Percent Change in Average Lifespan of songs on Hot 100 Charts by year")
plt.xlabel("Year")
plt.ylabel('Percent change in lifespan per year')
plt.axhline(y=0, xmin=0, xmax=1000, linewidth=1, color = 'k')
if SAVING: 
  plt.savefig("./Plots/pct_change_lifespan1.png")
plt.show()
```
